## Control Flow: Conditionals

Conditionals are statements in programming that let your code make decisions.
They allow Python to execute certain blocks of code only if a specified condition is True.

Think of it like everyday “if-then” logic:

“If it rains, then take an umbrella. Otherwise, wear sunglasses.”

### Types of Conditional Statements

Python has three main conditional statements:

### a) if

Executes a block of code only if the condition is True.

Syntax:

```python
if condition:
    # code block runs if condition is True
```

### b) else

- Executes a block of code if the if condition is False.

- Syntax:

```python

if condition:
    # runs if True
else:
    # runs if False
```


### c) elif (else if)

Used when there are multiple conditions to check.

Python checks conditions in order; the first True condition runs, the rest are skipped.

Syntax:

```python
if condition1:
    # code
elif condition2:
    # code
elif condition3:
    # code
else:
    # code if none above are True

```
