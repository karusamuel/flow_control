## Control Flow: Conditionals

Conditionals are statements in programming that let your code make decisions.
They allow Python to execute certain blocks of code only if a specified condition is True.

Think of it like everyday “if-then” logic:

“If it rains, then take an umbrella. Otherwise, wear sunglasses.”

### Types of Conditional Statements

Python has three main conditional statements:

### a) if

Executes a block of code only if the condition is True.

Syntax:

```python
if condition:
    # code block runs if condition is True
```

````python
print("starting...")
if True:
    print("i Run")
    print("hello world")
    print("am owesome")

print("The end!")
````

````python
canvas_hours = 11
attempted_checkpoint = True

if canvas_hours > 10:
    print("Good job")
````

````python
if canvas_hours > 10 and attempted_checkpoint:
    print("owesome")
````

### b) else

- Executes a block of code if the if condition is False.

- Syntax:

```python

if condition:
    # runs if True
else:
    # runs if False
```


````python
gender = None

if gender == "M":
    print("hello sir")
else:
    print("Hello madam")
````

### c) elif (else if)

Used when there are multiple conditions to check.

Python checks conditions in order; the first True condition runs, the rest are skipped.

Syntax:

```python
if condition1:
    # code
elif condition2:
    # code
elif condition3:
    # code
else:
    # code if none above are True

```

````python
gender = "M"

if gender == "M":
    print("Hello sir")
elif gender == "F":
    print("Hello Madam")
else:
    print("Hello",gender)
````

## Truthiness and Falsiness

Python allows any value to be treated as True or False in a conditional:

**Falsy values (considered False)**

    0 (zero)

    "" (empty string)

    [] (empty list)

    None (missing value)

**Truthy values (considered True)**

    Non-zero numbers

    Non-empty strings/lis

````python
# example

if  "":
    print("this action wont run")
````

````python
if  []:
    print("this action wont run")
````

````python
if  {}:
    print("this action wont run")
````

````python
if  None:
    print("this action wont run")
````

````python
if  0:
    print("this action wont run")
````

## Blocks and Indentation

Blocks are groups of code that belong together under a condition.

Indentation (4 spaces)(1 tab) is required to define a block.

````python
# example

if True: 
    #block start
    print("am here")
    print("am here")
    print("am here")
    print("am here")
    print("am here")
    print("am here")
    print("am here")
    print("am here")
    # block end
print("last") # not part of the block
````
